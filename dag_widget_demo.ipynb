## 1. Static view — Simple

```python
class ForeachDemoFlow(FlowSpec):
    @step
    def start(self):
        self.next(self.branch_a, self.branch_b)

    @step
    def branch_a(self):
        self.next(self.branch_f)
    
    @step
    def branch_b(self):
        self.items = ["A","B","C","D"]
        self.next(self.branch_c,foreach="items")
    @step
    def branch_c(self):
        self.next(self.branch_d,self.branch_e)
    @step
    def branch_d(self):
        self.next(self.join)
    @step
    def branch_e(self):
        self.next(self.join)
    @step
    def branch_f(self):
        self.next(self.join)
    @step
    def join(self, inputs):
        items = [1,2,3]
        self.next(self.end)

    @step
    def end(self):
        print("Done:", self.results)

DagWidget(ForeachDemoFlow)
```

## 2. Static view - Overflow Horizontal

```python
from metaflow import FlowSpec, step, NBRunner
from metaflow_jupyter import DagWidget
class BranchFlow(FlowSpec):
    @step
    def start(self):
        self.next(self.branch_a, self.branch_b, self.branch_c)

    @step
    def branch_a(self):
        self.result = "A"
        self.next(self.join)

    @step
    def branch_b(self):
        self.result = "B"
        self.next(self.join)
    @step
    def branch_c(self):
        self.result = "C"
        self.next(self.join)

    @step
    def join(self, inputs):
        self.results = [i.result for i in inputs]
        self.next(self.linear1)

    @step
    def linear1(self):
        self.next(self.linear2)
    @step
    def linear2(self):
        self.next(self.linear3)
    @step
    def linear3(self):
        self.next(self.linear4)
    @step
    def linear4(self):
        self.next(self.linear5)
    @step
    def linear5(self):
        self.next(self.linear6)
    @step
    def linear6(self):
        self.next(self.linear7)
    @step
    def linear7(self):
        self.next(self.end)

    @step
    def end(self):
        print(self.results)

DagWidget(BranchFlow)
```
